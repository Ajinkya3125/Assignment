```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes

df = load_diabetes()
```

```python
print(df["DESCR"])
```

```python
df_diabetes = pd.DataFrame(df.data,columns=["age","sex","bmi","bp","s1","s2","s3","s4","s5","s6"])
```

```python
df_diabetes.head()
```

```python
##Independent and dependent features
X = df_diabetes
y = df["target"]
```

```python
##train test split the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.head()
```

```python
##check the correlation
X_train.corr()
```

```python
#plot this by using heatmap
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

Hyperparameter Tunning

```python
param = {
    "criterion":["squared_error", "riedman_mse","absolute_error","poisson"],
    "max_depth":[1,2,3,4,5,10,15,20,25],
    "splitter":["best", "random"],
    "max_features":["auto","sqrt", "log2"]
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV

regressor = DecisionTreeRegressor()
grid = GridSearchCV(regressor,param_grid=param,cv=5,scoring="neg_mean_squared_error")
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
selectmodel = DecisionTreeRegressor(criterion='poisson',max_depth=4,max_features="sqrt",splitter="best")
```

```python
selectmodel.fit(X_train,y_train)
```
