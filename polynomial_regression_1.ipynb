```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
X = 6 * np.random.rand(100,1) - 3
y = 0.5 * X**2 + 1.5 * X + 2 + np.random.randn(100,1)
```

```python
plt.scatter(X,y,color="red")
plt.xlabel("X dataset")
plt.ylabel("Y dataset")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
##Lets implement linear regression
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,lr.predict(X_test))
print(score)
```

```python
##Lets visualize the dataset
plt.plot(X_train,lr.predict(X_train),color="red")
plt.scatter(X_train,y_train)
plt.xlabel("X dataset")
plt.ylabel("Y dataset")
```

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(degree=2,include_bias=True)
x_train_pf = pf.fit_transform(X_train)
x_test_pf = pf.transform(X_test)
```

```python
x_train_pf
```

```python
x_test_pf
```

```python
##
from sklearn.metrics import r2_score
lr = LinearRegression()
lr.fit(x_train_pf,y_train)
y_pred = lr.predict(x_test_pf)
score = r2_score(y_test,y_pred)
print(score)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.scatter(X_train,lr.predict(x_train_pf))
plt.scatter(X_train,y_train)
```

```python
##prediction of new dataset
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_pf = pf.transform(X_new)
X_new_pf
```

```python
y_new = lr.predict(X_new_pf)
plt.plot(X_new,y_new,"r-",linewidth=2,label="predictions")
plt.plot(X_train,y_train,"b.",label="Training points")
plt.plot(X_test,y_test,"g.",label="Testing points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

Pipeline concepts

```python
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
def poly_regression(degree):
    X_new = np.linspace(-3,3,200).reshape(200,1)
    poly_features = PolynomialFeatures(degree=degree,include_bias=True)
    lr = LinearRegression()
    poly_regression = Pipeline([
        ("poly_features",poly_features),
        ("Linear regression",lr)
    ])
    poly_regression.fit(X_train,y_train)

    y_pred_new = poly_regression.predict(X_new)
    ##plotting prediction line
    plt.plot(X_new,y_pred_new,"r",label="Degree " + str(degree), linewidth=2)
    plt.plot(X_train,y_train,"b.",linewidth=3)
    plt.plot(X_test,y_test,"g.",linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
poly_regression(5)
```
