```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
print(iris["target"])
```

```python
print(iris["DESCR"])
```

```python
##Independent feature
X = pd.DataFrame(iris["data"],columns=["sepal lenght","sepal width","petal length","petal width"])
```

```python
##dependent feature
y = iris["target"]
```

```python
##train test split the dataset
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
##apply the decision tree classifier algorithm
from sklearn.tree import DecisionTreeClassifier
treeclass = DecisionTreeClassifier()
treeclass.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(11,8))
tree.plot_tree(treeclass,filled=True)
```

```python
##prediction
y_pred = treeclass.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print("Accuracy of the model:",accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

Decision Tree Classifier PrePruning and Hyperparameter

```python
param = {
    "criterion":["gini", "entropy", "log_loss"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5],
    "max_features":["sqrt","log2","auto"]
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV

treeclass = DecisionTreeClassifier()
grid = GridSearchCV(treeclass,param_grid=param,cv=5,scoring="accuracy")
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
