```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv("cardekho_imputated.csv",index_col=[0])
df.head()
```

```python
##check null values
##check features with NaN values
df.isnull().sum()
```

```python
##Remove unnecessary columns
df.drop('car_name',axis=1,inplace=True)
df.drop('brand',axis=1,inplace=True)
```

```python
df.head()
```

```python
df["model"].unique()
```

```python
##getting all different types of features
num_features = [feature for feature in df.columns if df[feature].dtype != 'O']
print("Number of numerical features:",len(num_features))

cat_features = [feature for feature in df.columns if df[feature].dtype == 'O']
print("Number of categorical features:",len(cat_features))

discrete_features = [feature for feature in num_features if len(df[feature].unique()) <= 25]
print("Number of continuous features:",len(discrete_features))

continuous_features = [feature for feature in num_features if feature not in discrete_features]
print("Number of categorical features:",len(continuous_features))
```

```python
##Independent and dependent features
from sklearn.model_selection import train_test_split
X = df.drop(["selling_price"],axis=1)
y = df["selling_price"]
```

```python
X.head()
```

```python
len(df["model"].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X["model"] = le.fit_transform(X["model"])
```

```python
X.head()
```

```python
len(df["seller_type"].unique()),len(df["fuel_type"].unique()),len(df["transmission_type"].unique())
```

```python
##create column transformer with 3 types of transformer
num_features = X.select_dtypes(exclude="object").columns
onehot_columns = ["seller_type","fuel_type","transmission_type"]

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop="first")

preprocessor = ColumnTransformer(
    [
        ("OneHotEncoder",oh_transformer,onehot_columns),
        ("StandardScaler",numeric_transformer,num_features)
    ],remainder="passthrough"
)
```

```python
X = preprocessor.fit_transform(X)
```

```python
pd.DataFrame(X)
```

```python
##seperate dataset into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
X_train.shape, X_test.shape
```

```python
X_train
```

##Model training and model selection

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
##create a function to evaluate a model
def evaluate_model(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    rmse = np.sqrt(mean_squared_error(true,predicted))
    r2_square = r2_score(true,predicted)
    return mae,rmse,r2_square
```

```python
##Beginning model training
models = {
    "Random Forest Regressor":RandomForestRegressor()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)##train model

    #Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    #Evaluate Train and Test dataset
    model_train_mae,model_train_rmse,model_train_r2 = evaluate_model(y_train,y_train_pred)
    model_test_mae,model_test_rmse,model_test_r2 = evaluate_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    print("Model Performance For Training set")
    print("- Root Mean Squared Error:{:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error:{:.4f}".format(model_train_mae))
    print("- R2 Score:{:.4f}".format(model_train_r2))

    print("-------------------------------------------------------")

    print("Model Performance For Testing set")
    print("- Root Mean Squared Error:{:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error:{:.4f}".format(model_test_mae))
    print("- R2 Score:{:.4f}".format(model_test_r2))

    print("="*35)
    print("\n")
```

```python
##initialize few parameter for Hyperparameter tuning
rf_params = {"max_depth":[5,8,15,None,10],
            "max_features":[5,7,"auto",8],
            "min_samples_split":[2,8,15,20],
            "n_estimators":[100,200,500,1000]
            }
```

```python
#Models List for hyperparameter tuning
randomcv_models = [
                    ("RF",RandomForestRegressor(),rf_params)
                ]
```

```python
##Hyperparameter Tuning
from sklearn.model_selection import RandomizedSearchCV
model_param = {}
for name,model,params in randomcv_models:
    random = RandomizedSearchCV(estimator=model,
                                param_distributions=params,
                                n_iter=100,
                                cv=3,
                                verbose=2,
                                n_jobs=-1)
    random.fit(X_train,y_train)
    model_param[name] = random.best_params_

for model_name in model_param:
    print(f"-------------------------Best Params for {model_name} ----------------------------")
    print(model_param[model_name])
```

```python
##Retraining the models with best parameters
models = {
    "Random Forest Regressor":RandomForestRegressor(n_estimators=100,min_samples_split=2,max_features=7,max_depth=None,n_jobs=-1)
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)##train model

    #Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    #Evaluate Train and Test dataset
    model_train_mae,model_train_rmse,model_train_r2 = evaluate_model(y_train,y_train_pred)
    model_test_mae,model_test_rmse,model_test_r2 = evaluate_model(y_test,y_test_pred)

    print(list(models.keys())[i])
    print("Model Performance For Training set")
    print("- Root Mean Squared Error:{:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error:{:.4f}".format(model_train_mae))
    print("- R2 Score:{:.4f}".format(model_train_r2))
    
    print("-------------------------------------------------------")

    print("Model Performance For Testing set")
    print("- Root Mean Squared Error:{:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error:{:.4f}".format(model_test_mae))
    print("- R2 Score:{:.4f}".format(model_test_r2))
    
    print("="*35)
    print("\n")
```

```python
threshold = y_train.median()
y_train_cls = (y_train >= threshold).astype(int)
y_test_cls = (y_test >= threshold).astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
def evaluate_classification(true, predicted):
    accuracy = accuracy_score(true, predicted)
    precision = precision_score(true, predicted)
    recall = recall_score(true, predicted)
    return accuracy, precision, recall
```

```python
models = {
    "Random Forest Classifier": RandomForestClassifier(random_state=42)
}

for i in range(len(models)):
    model = list(models.values())[i]
    model.fit(X_train, y_train_cls)

    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluation
    train_acc, train_prec, train_rec = evaluate_classification(y_train_cls, y_train_pred)
    test_acc, test_prec, test_rec = evaluate_classification(y_test_cls, y_test_pred)

    print(list(models.keys())[i])
    print("Training Set Performance")
    print(f"- Accuracy: {train_acc:.4f}")
    print(f"- Precision: {train_prec:.4f}")
    print(f"- Recall: {train_rec:.4f}")

    print("------------------------------------------------")

    print("Testing Set Performance")
    print(f"- Accuracy: {test_acc:.4f}")
    print(f"- Precision: {test_prec:.4f}")
    print(f"- Recall: {test_rec:.4f}")

    print("=" * 35, "\n")
```
